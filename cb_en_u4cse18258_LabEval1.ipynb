```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
df = pd.read_csv('heart.csv')

print(df)
```

Checking if for any NULL values


```
print("No of NULL values in the dataset ",len(df)-len(df.isna()))
```

Removing random NULL values


```
for col in df.columns:
    df.loc[df.sample(frac=0.2).index, col] = np.nan
```

```
df
```

Removing NULL values from the dataset



```
for col in df.columns:
  df[col].fillna(df[col].mode()[0], inplace=True)
```

```
df
```

Computing Mean


```
print(df.mean())
```

Computing Median


```
print(df.median())
```

Computing Mode

```
for col in df.columns:
  print("Mode of column ",col," is",df[col].mode()[0])
```

Range


```
for col in df.columns:
  print(col,max(df[col])-min(df[col]))
```

IQR

```
Q1 = np.percentile(df, 25, interpolation = 'midpoint')
Q3 = np.percentile(df, 25, interpolation = 'midpoint')

IQR = Q3 - Q1

print(IQR)
print(Q1)
print(Q3)
```

Plots Shown along with the labels

```
plt.figure(figsize= (20,7))
plt.style.use('ggplot')
sns.lineplot(df['age'],df['cp'],linewidth=2,hue=df['sex'],ci=None,palette=['blue','yellow'])
plt.xlabel("Age",fontsize=20)
plt.ylabel("Chest Pain Type",fontsize=20)
plt.title("Age versus Chest Pain Type",fontsize=25)
plt.legend(labels=['Male','Female'],frameon=True,fontsize='large',bbox_to_anchor=(1.05, 0.2),loc=2)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.style.use('ggplot')
sns.lineplot(df["cp"],df['chol'],linewidth=2,hue=df['sex'],ci=None,palette=['purple','yellow'])
plt.xlabel("Chest Pain",fontsize=20)
plt.ylabel("Cholestrol",fontsize=20)
plt.title("Chest Pain versus Cholestrol",fontsize=25)
plt.legend(labels=['Male','Female'],frameon=True,fontsize='large',bbox_to_anchor=(1.05, 0.2),loc=2)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.style.use('ggplot')
sns.lineplot(df["age"],df['thal'],linewidth=2,hue=df['sex'],ci=None,palette=['yellow','purple'])
plt.xlabel("Age",fontsize=20)
plt.ylabel("Thal",fontsize=20)
plt.title("Age versus Thal",fontsize=25)
plt.legend(labels=['Male','Female'],frameon=True,fontsize='large',bbox_to_anchor=(1.05, 0.2),loc=2)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.style.use('ggplot')
sns.lineplot(df["sex"],df['cp'],linewidth=2,hue=df['target'],ci=None,palette=['red','green'])
plt.xlabel("Sex",fontsize=20)
plt.ylabel("Chest Pain",fontsize=20)
plt.title("Sex versus Chest Pain",fontsize=25)
plt.legend(labels=['Not Heart Attack','Heart Attack'],frameon=True,fontsize='large',bbox_to_anchor=(1.05, 0.2),loc=2)
plt.show()
```
